```python
import pandas as pd
```

```python
df = pd.read_csv("data/archive/imdb_top_1000_cleaned.csv")
df.head()
```

```python
df.drop(df[df["Released_Year"] == "PG"].index,inplace=True)
```

```python
def func(year):
    if year in range(1920,1930):
        return "1920s"
    elif year in range(1930,1940):
        return "1930s"
    elif year in range(1940,1950):
        return "1940s"
    elif year in range(1950,1960):
        return "1950s"
    elif year in range(1960,1970):
        return "1960s"
    elif year in range(1970,1980):
        return "1970s"
    elif year in range(1980,1990):
        return "1980s"
    elif year in range(1990,2000):
        return "1990s"
    elif year in range(2000,2010):
        return "2000s"
    elif year in range(2010,2020):
        return "2010s"
    elif year == 2020:
        return "2020"

# df["Range"] = df["Released_Year"].apply(func)
```

```python
df["Released_Year"] = pd.to_numeric(df["Released_Year"])
df["range"] = df["Released_Year"].apply(func)
```
